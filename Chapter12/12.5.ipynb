# 12.5 Equations of Lines and Planes

A line $L$ in three-dimensional space is determined when we know a point $P_0(x_0,y_0,z_0)$ on $L$ and the direction of $L$.
![Line in space](./line_in_space.png)

## Vector Equation of $L$
$$
{\bf r} = {\bf r_0} + t{\bf v}
$$

## Parametric Equations of $L$
$$
x=x_0+at, \quad y=y_0+bt, \quad z=z_0+ct 
$$

$a,b, c$ are called __direction numbers__ of $L$.

## Symmetic Equations of $L$
$$
\frac{x-x_0)}{a}= \frac{y-y_0)}{b}= \frac{z-z_0)}{c}
$$

### Example
* Find a vector equation, parametric equations and symmetric equations of the line that passes through the points $A(2,4,-3)$ and $B(3,-1,1)$.

# Planes
## Vector Equation of the Plane
$$
{\bf n}\cdot( {\bf r}-{\bf r_0})=0
$$
${\bf n}$ is called __normal vector__ on (of) the plane.
## A Scalar equation of the plane
through the point $P_(x_0,y_0,z_0)$ with normal vector ${\bf n}=\langle a,b,c\rangle $ is 
$$
a(x-x_0)+b(y-y_0)+c(z-z_0)=0
$$

## Distances
Find a formula for the distance $D$ from a point $P_1(x_1,y_1,z_1)$ to the plane $ax+by+cz+d=0$.
$$
D= \frac{\left|ax_1+by_1+cz_1+d\right|}{\sqrt{a^2+b^2+c^2}}.
$$
