## $\Delta r = (b-a) /m$, $\Delta \theta = (\beta-\alpha)/n$
![Polar Integrals](./15-3-e.png)

$$
R_{ij} = \left\{(r, \theta) \;|\; r_{i-1}\leq r \leq r_i, \theta_{j-1} \leq \theta \leq \theta_j \right\}
$$

center point $(r^*_i, \theta_j^*)$
$$
r^*_i={1\over 2}(r_{i-1}+r_i) , \qquad \theta_j^*={1\over 2}(\theta_{j-1}+\theta_j)
$$

Now 
$$
\Delta A_i = {1\over 2}r_i^2\Delta \theta - {1\over 2}r_{i-1}^2\Delta \theta =
{1\over 2}(r_{i}^2-r_{i-1}^2)\Delta \theta = {1\over 2}(r_{i}+r_{i-1})(r_{i}-r_{i-1})\Delta \theta = r^*_i \Delta r \Delta \theta
$$

So,
$$
\sum_{i=1}^m \sum_{j=1}^n f(r^*_i \cos \theta^*_j, r^*_i \sin \theta^*_j) \Delta A_i =\sum_{i=1}^m \sum_{j=1}^n f(r^*_i \cos \theta^*_j, r^*_i \sin \theta^*_j) r^*_i \Delta r \Delta \theta
$$
Let 
$$
g(r,\theta) = r f(r\cos \theta, r\sin\theta) 
$$
Hence 
$$
\begin{array}{lll}
\iint_R f(x,y) dA &=& \lim_{m,n\to \infty} \sum_{i=1}^m \sum_{j=1}^n f(r^*_i \cos \theta^*_j, r^*_i \sin \theta^*_j) \Delta A_i  \\
  &=& \lim_{m,n\to \infty} \sum_{i=1}^m \sum_{j=1}^n f(r^*_i \cos \theta^*_j, r^*_i \sin \theta^*_j) r^*_i \Delta r \Delta \theta \\
  &=& \lim_{m,n\to \infty} \sum_{i=1}^m \sum_{j=1}^n g(r^*_i , \theta^*_j) \Delta r \Delta \theta \\
  &=& \int_{\alpha}^{\beta}\int_a^b g(r,\theta) dr d\theta \\
  &=& \int_{\alpha}^{\beta}\int_a^b f(r\cos\theta, r\sin\theta) r dr d\theta 
\end{array}
$$

![Polar Integrals](./15-3-f.png)
