## Properties of Double Integrals
1. $$ \iint_D\left[f(x,y) + g(x,y) \right] dA =\iint_D f(x,y) dA + \iint_D g(x,y) dA $$
2. $$ \iint_D c\left[f(x,y)\right] dA =c \iint_D f(x,y) dA $$
3. If $f(x,y)\geq g(x,y)$ for all $(x,y) \in D$, then
$$
 \iint_D f(x,y) dA \geq \iint_D g(x,y) dA 
$$
4. ![Prop 8](./15-2-g.png)
$$
 \iint_D f(x,y) dA = \iint_{D_1} f(x,y) dA +  \iint_{D_2} f(x,y) dA
$$
5. 
$$
\iint_{D} 1 dA = A(D)
$$
6. If $m\leq f(x) \leq M$ for all $(x,y) \in D$, then
$$
mA(D) \leq \iint_D f(x,y) dA \leq MA(D)
$$
