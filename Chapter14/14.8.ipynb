# 14.8 Lagrange Multipliers

Consider the problem 
\begin{array}{rl}
\min & f(x,y)\\
\text{subject to} & \\
 & g(x,y)=k
\end{array}

### Example
\begin{array}{rl}
\min & x^2+y^2\\
\text{subject to} & \\
 & xy=1
\end{array}

### Remark 
Suppose that $f$ has an extreme value at $(x_0, y_0, z_0)$. Then
$$
\nabla f(x_0, y_0, z_0) = \lambda \nabla g(x_0, y_0, z_0)
$$
The number $\lambda$ in Equation 1 is called a Lagrange multiplier.

## Two Constraints
\begin{array}{rl}
\min & f(x,y)\\
\text{subject to} & \\
 & g(x,y)=k
 & h(x,y)=c
\end{array}

### Lagragne Methods
(under certain conditions there exist two numbers $\lambda$ and $\mu$ such that)
$$
\nabla f(x_0,y_0,z_0) = \lambda \nabla g(x_0,y_0,z_0) + \mu \nabla h(x_0,y_0,z_0) 
$$

### Example
Find the maximum value of the function $f(x,y,z)=x+2y+3z$ on the curve of the intersection of the plane $x-y+z=1$ and the cylinder $x^2+y^2=1$
