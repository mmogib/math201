## Functions of two variables

![Definition](./fun_def.png)

* we write $z = f(x,y)$
* $x,y$ are __independent variables__
* $z$ is __dependent variables__

### Example

1. $f(x,y)=\frac{\sqrt{x+y+1}}{x-1}$
2. $f(x,y)=x\ln(y^2-x)$

## Functions of Three or More Variables
A function of three variables, $f$, is a rule that assigns to each ordered triple $(x, y, z)$ in a domain $D \subset \mathbb{R}^3$ a unique real number denoted by $f(x, y, z)$.

### Example
Find the domain of $f$ if
$$
f(x,y,z) = \ln(z-y) + xy\sin z
$$

### Example 
Find the level surfaces of
$$
f(x,y,z) = x^2+y^2+z^2
$$
