## Linear Approximations
The linear function whose graph is the tangent plane of $z=f(x,y)$ at the point $(x,y,f(x,y))$; that is
$$
L(x,y) = f(a,b) +f_x(a,b)(x-a)+f_y(a,b)(y-b)
$$
is called __linear approximation__ or __tangent plane approximation__ of $f$ at $(a,b)$

## Differentials
For a differentiable function of two variables, $z = f (x, y)$, 

* we define the differentials $dx$ and $dy$ to be independent variables; 
* Then the differential $dz$, also called the __total differential__, is defined by 
$$
dz = f_x(x,y)dx+f_y(x,y)dy = \frac{\partial z}{\partial x}x + \frac{\partial z}{\partial y}y
$$


### Example

2. If $x$ changes from $2$ to $2.05$ and $y$ changes from $3$ to 	$2.96$, compare the values of $\Delta z$ and $dz$.
